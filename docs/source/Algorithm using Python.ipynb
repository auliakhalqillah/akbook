## If and Else (Logic)

There is a logic process in programming that has two or more conditions, such as `yes` and `no` or also like categorical process such as `low`, `medium`, and `high`. The logic process applies some conditions:

|condition|sign|
|:---------|:----:|
|Less than|`<`|
|Less than and equal to|`<=`|
|Greater than|`>`|
|Greater than and equal to|`>=`|
|Equal to|`==`|
|Not equal to|`!=`|

The implimentation of simple logic process in Python as follows


```python
a = 10
b = 60
if a < b:
  print(f"{a} is less than {b}")
else:
  print(f"{a} is greater than {b}")
```

for the simple categorical proces in Python can be written as follows


```python
hot_day = 30 # in degree Celsius
cold_day = 10 # in degree Celsius


current_temperature = 22 # in degree Celsius

if current_temperature > hot_day:
  print("It's a hot day")
elif current_temperature >= cold_day:
  print("It's a normal day")
else:
  print("It's a cold day")
```

## Indexing

In Python, a list of data has its index number. The index number is started from 0. For an example, if we have a list of data as `data = [2,1,8]`, then the first data will be indexed as 0, the second one is indexed as 1 and so on. To assign the index, we use the square brackets `[]`. Here is the exmaple



```python
data = [2,1,8]
print(f"Index 0 = {data[0]}")
print(f"Index 1 = {data[1]}")
print(f"Index 2 = {data[2]}")
```

We notice that the number of data is three, but the index number will be end at two (2). Therefore, the index number in Python has a pattern as `0:N-1`. The index can be accessed in list and tuple data types.

## Looping (Iterating)

Looping or iterating is one of the basic algorithm in programming. The same process will be repeated until the maximum number of iteration has reached. This process will help us if there are a lot of datasets that we should proccess at one go. In Python, we can use two main looping approach, `for` and `while` methods.





### For Method

```python
for i in range(5):
  print(i)
```

The code above is the standard code to iterate the process. The `range(5)` means that there are 5 data will be repeated and `i` is the index indicator for each looping. We can try by using the variabel of `data` that has been defined previously.

```python
n = len(data)
for i in range(n):
  print(data[i])
```

The number of data is defined as variabel of `n` by using `len(data)` function. Then the variabel of `n` is called in the looping process and we print for each data based on the index indicator `i`. Everything code inside the `for` will be repeated.

### While Method

```python
i = 0
n = 5
while i < n:
  print(i)
  i = i + 1
```

Looping by uisng `while` method applies logic term (`<` or `>`) in the process. First, we have to define the index indicator (by ourself) `i` and number of data `n`. We then need update the `i` for the next iteration through the code of `i = i + 1`. Since the initial `i` is 0, then the next `i` is `i = 0 + 1 = 1`.

## An Example of Simple Algorithm

Given a student values for a certain subjuect, calculate the average value of all students and define the categorical value based on the given indicator as below!

__Student Values__

|Student Name|Value|
|:-----------|:---:|
|Angle|60|
|Baron|75|
|Charlie|40|
|Donie|98|
|Emma|65|



__Category Values__

|Value|Category|
|:-----------|:---:|
|>=87|A|
|>=77|B|
|>=67|C|
|>=57|D|
|< 57|E|



```python
student_names = ["Angle","Baron","Charlie","Donie","Emma"]
student_values = [60,75,40,98,65]
n_student = len(student_values)

# sum all values
total = 0
for i in range(n_student):
  total = total + student_values[i]

print(f"Number of Student = {n_student}")
print(f"Total value = {total}")

# calculate average value
average_value = total / n_student
print(f"Average value = {average_value}")

# Check the category value
cat_value = list()
for i in range(n_student):
  if student_values[i] >= 87:
    cat_value.append("A")
  elif student_values[i] >= 77:
    cat_value.append("B")
  elif student_values[i] >= 67:
    cat_value.append("C")
  elif student_values[i] >= 57:
    cat_value.append("D")
  else:
    cat_value.append("E")

# print category value for each student name
for i in range(n_student):
  print(f"{student_names[i]} = {student_values[i]} ({cat_value[i]})")
# print(f"Category value = {cat_value}")
```
